```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr_matrix = np.random.random((10,10))
corr_matrix = corr_matrix*2 -1
mask = np.triu(np.ones_like(corr_matrix,dtype=bool))
plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix, mask=mask, center=0,  cmap='coolwarm', vmax=1.2, vmin=-1.2, cbar=False, xticklabels=False, yticklabels=False)  #annot=True,显示中间的数字
```

```python
# mask2 = np.abs(corr_matrix)<0.5
mask3 = mask | mask2
plt.figure(figsize=(8,8))
sns.heatmap(corr_matrix, mask=mask3, center=0,  cmap='coolwarm', vmax=1.2, vmin=-1.2, cbar=False, xticklabels=False, yticklabels=False)  #annot=True,显示中间的数字
```
